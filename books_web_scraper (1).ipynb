```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import lxml.html
from urllib.parse import urldefrag
```

```python
title_list=[]
availability_list=[]
price_list=[]
url_list=[]

for page in range(1,51):
  html_text=requests.get(f"https://books.toscrape.com/catalogue/page-{page}.html").content
  soup=BeautifulSoup(html_text, 'lxml')

  #part1
  part1=soup.find_all('li', class_='col-xs-6 col-sm-4 col-md-3 col-lg-3')
  for p in part1:
    for a in p.find_all('a'):
      title=a
    title=title['title'] 
    title_list.append(title)
    availability = p.find('p', class_='instock availability').text.replace('\n','').replace(' ','')
    availability_list.append(availability)
    price=p.find('p', class_='price_color').text
    price_list.append(price)

  #part2
  part2=soup.find_all('article')
  for p in part2:
    for a in p.find_all('a', href=True):
      url=a['href']
    url_list.append('https://books.toscrape.com/catalogue/'+url) 
  

#part3
ups_list=[]
category_list=[]
for url in url_list:  
  html_text=requests.get(url).content
  soup_new=BeautifulSoup(html_text, 'lxml')
  ups=soup_new.find('td').text
  ups_list.append(ups)
  category=soup_new.find_all('li')[2].text.replace('\n','')
  category_list.append(category)
```

```python
df=pd.DataFrame({
    'title':title_list,
    'availability':availability_list,
    'price':price_list,
    'url':url_list,
    'ups':ups_list,
    'category':category
})

df.to_csv('books_data.csv')
```
